# Sliding Fall data 

### Sliding window function : 
Parameters : 

    W=200 : size of sliding window 
    
    overlap =50% = (200/2)=100

```python
import os
import scipy.io
import pandas as pd
import matplotlib.pyplot as plt
import mat4py as mp
import numpy as np
dataDir = "C:/Users/azerty/Desktop/Telnet/on-wirst Dataset/trainingSetAccGyroMag/Fall/mat format/"
for file in os.listdir( dataDir ) :
    file_name=os.path.basename(dataDir+file)[0:6]
    fall=[]
    fall.append(scipy.io.loadmat( dataDir+file ) )

    h=fall[0]['TA']


    h=pd.DataFrame(h)
    h=h.transpose()

    windowList_fall=[]

    def sliding_window(dataframe):           
        w = 400 # Size of sliding window     
        for i in range(0, len(dataframe)-200,100):
            w1 = dataframe.iloc[i:i+w][:] 
            windowList_fall.append(w1)

    

    sliding_window(h)  
    print(windowList_fall)

    windowList_fall=pd.DataFrame(windowList_fall)

    windowList_fall.to_csv("C:/Users/azerty/Desktop/Telnet/on-wirst Dataset/trainingSetAccGyroMag/Fall/Sliding TA csv/"+file_name+'.csv',header="TA Slicing")

    
```
